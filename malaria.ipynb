```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path='/content/drive/MyDrive/mldata0.xlsx'
df = pd.read_excel(file_path, sheet_name=7)
df
```

```python
df.columns
```

```python
plt.figure(figsize=(7, 7))
plt.pie(df["PositiveCases"], labels=df["HealthRegion"], autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Proportion of Malaria Positive Cases by Health Region")
plt.show()

```

```python


fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.2
index = np.arange(len(df["HealthRegion"]))


plt.bar(index, df['P.falciparum(Malignant)'], bar_width, label="P. falciparum (Malignant)")
plt.bar(index + bar_width, df["P. vivax and Povale (Benign Tertiary)"], bar_width, label="P. vivax and Povale (Benign Tertiary)")
plt.bar(index + 2*bar_width, df["P. malariae (Benign Quartan)"], bar_width, label="P. malariae (Benign Quartan)",color="brown")
plt.bar(index + 3*bar_width, df["Mixed"], bar_width, label="Mixed")

# Formatting the graph
plt.xlabel("Health Region")
plt.ylabel("Number of Cases")
plt.title("Distribution of Malaria Types by Health Region")
plt.xticks(index + 1.5 * bar_width, df["HealthRegion"], rotation=30)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)


plt.show()
```
