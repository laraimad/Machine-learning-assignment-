```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path='/content/drive/MyDrive/mldata0.xlsx'
```

```python
df = pd.read_excel(file_path, sheet_name=3)
df.head()
```

```python

# Select relevant columns for analysis
df_filtered = df.iloc[:, 1:]  # Remove disease names column
df_filtered.head()
```

```python


# Extract relevant columns (drop disease names)
df_filtered = df.iloc[:, 2:]


for col in df_filtered.columns:
    df_filtered[col] = pd.to_numeric(df_filtered[col], errors='coerce')

# Sum cases for each category
young_cases = df_filtered.iloc[:, :3].sum().sum()  # Sum first 3 columns (children)
older_cases = df_filtered.iloc[:, 3:].sum().sum()  # Sum last 2 columns (adults)

# The a bar chart
categories = ["Children (Young)", "Older Adults"]
values = [young_cases, older_cases]

plt.figure(figsize=(8,5))
plt.bar(categories, values, color=["blue", "orange"])
plt.xlabel("Age Group")
plt.ylabel("Total Infection Cases")
plt.title("Comparison of Infection Cases: Children vs. Older Adults")
plt.show()
```
